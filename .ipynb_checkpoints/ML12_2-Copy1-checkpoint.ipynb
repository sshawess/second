```python
def getsearch(pageviewid, dataframe):
    #for givien page view id find max page view  in dataframe with onsitesearchterm not null and pageviewid 
    #is less than or equal to given pageviewid
    try:
        termpageviewid=max(i for i in dataframe['PAGE_VIEW_ID'] 
        if i<=pageviewid
        and dataframe['ON_SITE_SEARCH_TERM'].loc[dataframe['PAGE_VIEW_ID']== i].values[0] is not None
    
        and dataframe['SESSION_ID'].loc[dataframe['PAGE_VIEW_ID']==i].values[0]
        ==dataframe['SESSION_ID'].loc[dataframe['PAGE_VIEW_ID']==pageviewid].values[0])       
                       
    #and dataframe['SESSION_ID'].loc[dataframe['PAGE_VIEW_ID']==i] == 
    #dataframe['SESSION_ID'].loc[dataframe['PAGE_VIEW_ID']==int(pageviewid)])
    
        return dataframe['ON_SITE_SEARCH_TERM'].loc[dataframe['PAGE_VIEW_ID']==termpageviewid].values[0]
    
    except:
        return 'no search'
```
